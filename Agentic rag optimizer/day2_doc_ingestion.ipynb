```python
#step 1 importing all libraries
from langchain.text_splitter import RecursiveCharacterTextSplitter
import os
import pandas as pd 
from PyPDF2 import PdfReader
from docx import Document
print("all libraries are ready")
```

```python
import os
if not os.path.exists("data"):
    os.makedirs("data")
    print("created 'data/' folder.")
else:
    print("'data/' folder already exists.")
```

```python
# step 2 :--functions to read different document types
def read_txt(file_path):
    with open(file_path, 'r', encoding='utf-8')as f:
        return f.read()

def read_pdf(file_path):
    reader = PdfReader(file_path)
    text = ""
    for page in reader.pages:
        text += page.extract_text() + "\n"
    return text

def read_docx(file_path):
    doc = Document(file_path)
    return "\n".join([para.text for para in doc.paragraphs])
```

```python
# step 3:--initialize text splitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=50,#charactes per chunk
                                               chunk_overlap=5)#overlap between chunks
```

```python
# step4:-- Load all documents from data folder to split into chunks
all_chunks = []

for file in os.listdir("data"):
    file_path = os.path.join("data", file)

    if file.endswith(".txt"):
        text = read_txt(file_path)
    elif file.endswith(".pdf"):
        text = read_pdf(file_path)
    elif file.endswith(".docx"):
        text = read_docx(file_path)
    else:
        continue
    chunks = text_splitter.split_text(text)
    for i,chunk in enumerate(chunks):
        all_chunks.append({"source": file,
                           "chunk_id": i,
                           "text": chunk })

# step 5:-- create dataframe and inspect
df_chunks = pd.DataFrame(all_chunks)
print("first 5 chunks")
print(df_chunks.head())

# step 6:-- save the chunks to csv for day 3
df_chunks.to_csv("data/chunks_day2.csv", index=False)
print("chunks saved to 'data/chunks_day2.csv'")
```
